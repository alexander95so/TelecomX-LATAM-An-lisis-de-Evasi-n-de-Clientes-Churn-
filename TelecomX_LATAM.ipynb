```python
columnas = ['customer','phone','internet','account']
```

```python
for columna in columnas:
  datos = datos.join(pd.json_normalize(datos.pop(columna)))
```

```python


datos.head()
```

```python


pd.set_option('display.max_columns', None)
datos.tail()
```

```python
import numpy as np
```

```python

datos.info()
```

```python
datos['Charges.Total'].eq(' ').sum()
```

```python


datos['Charges.Total'] = datos['Charges.Total'].replace(' ',0)
```

```python
import numpy as np
datos['Charges.Total'] = datos['Charges.Total'].astype(np.float64)
```

```python


datos.info()
```

```python
datos.head()
```

```python

datos['Cuentas_Diarias'] = datos['Charges.Monthly']/30
```

```python
datos.head()
```

```python


datos['Churn'] = (datos['Churn'] == "Yes").astype(np.int64)
datos['Partner'] = (datos['Partner'] == "Yes").astype(np.int64)
datos['Dependents'] = (datos['Dependents'] == "Yes").astype(np.int64)
datos['PhoneService'] = (datos['PhoneService'] == "Yes").astype(np.int64)
datos['OnlineSecurity'] = (datos['OnlineSecurity'] == "Yes").astype(np.int64)
datos['OnlineBackup'] = (datos['OnlineBackup'] == "Yes").astype(np.int64)
datos['DeviceProtection'] = (datos['DeviceProtection'] == "Yes").astype(np.int64)
datos['TechSupport'] = (datos['TechSupport'] == "Yes").astype(np.int64)
datos['StreamingMovies'] = (datos['StreamingMovies'] == "Yes").astype(np.int64)
datos['PaperlessBilling'] = (datos['PaperlessBilling'] == "Yes").astype(np.int64)
datos.info()
```

```python
datos.head()
```

```python
datos.to_csv('datos_tratados.csv')
```

#📊 Carga y análisis

```python


datos.describe()
```

```python
import matplotlib.pyplot as plt
```

```python


datos['Churn'].value_counts(normalize=True).plot(kind='pie',figsize=(8,6),autopct='%1.1f%%')
plt.title('Distribución de Evasión (Churn)')
plt.xlabel('Churn (0 = Permanece, 1 = Se dió de baja)')
plt.ylabel('Cantidad de clientes')
plt.show()
```

```python

evasion_genero = datos.groupby('gender')['Churn'].mean()

evasion_contrato = datos.groupby('Contract')['Churn'].mean()


evasion_pago = datos.groupby('PaymentMethod')['Churn'].mean()


evasion_partner = datos.groupby('Partner')['Churn'].mean()

evasion_senior = datos.groupby('SeniorCitizen')['Churn'].mean()


evasion_dependents = datos.groupby('Dependents')['Churn'].mean()


evasion_movies = datos.groupby('StreamingMovies')['Churn'].mean()
```

```python


evasion_genero.plot(kind='bar',figsize=(6,4))
plt.title('Distribución de Evasión - Por el género del cliente')
plt.ylabel('Proporción de Churn')
plt.xlabel('Género')
plt.show()
```

```python
evasion_contrato.plot(kind='bar',figsize=(6,4), color='r')
plt.title('Distribución de Evasión - Por el tipo de Contrato que tiene')
plt.ylabel('Proporción de Churn')
plt.xlabel('Tipo de Contrato')
plt.show()
```

```python
evasion_pago.plot(kind='bar',figsize=(6,4),color='b')
plt.title('Distribución de Evasión - Por el Método de Pago que usa')
plt.ylabel('Proporción de Churn')
plt.xlabel('Método de Pago')
plt.show()
```

```python
evasion_partner.plot(kind='bar',figsize=(6,4),color='y')
plt.title('Distribución de Evasión - Por si tiene o no pareja')
plt.ylabel('Proporción de Churn')
plt.xlabel('0 = Sin pareja; 1 = Con pareja')
plt.show()
```

```python
evasion_senior.plot(kind='bar',figsize=(6,4),color='purple')
plt.title('Distribución de Evasión - Por si es Senior Citizen o no')
plt.ylabel('Proporción de Churn')
plt.xlabel('0 = No es Senior Citizen; 1 = Es Senior Citizen (tiene >= 65 años)')
plt.show()
```

```python
evasion_dependents.plot(kind='bar',figsize=(6,4),color='black')
plt.title('Distribución de Evasión - Por si el cliente tiene o no dependientes')
plt.ylabel('Proporción de Churn')
plt.xlabel('0 = No tiene dependientes; 1 = Tiene dependientes')
plt.show()
```

```python
evasion_movies.plot(kind='bar',figsize=(6,4),color='g')
plt.title('Distribución de Evasión - Por si tiene una suscripción de\n Streaming de peliculas')
plt.ylabel('Proporción de Churn')
plt.xlabel('0 = No tiene una suscripción; 1 = Si tiene una suscripción')
plt.show()
```

```python

datos.head()
```

```python

evasion_total = datos.groupby('Churn')['Charges.Total'].mean()


evasion_tiempo_contrato = datos.groupby('Churn')['tenure'].mean()
```

```python
plt.figure(figsize=(6,4))

evasion_total.plot(kind='bar', color = 'brown')

plt.title('Gasto total promedio según evasión')
plt.xlabel('Churn (0 = Permanece, 1 = Se dio de baja)')
plt.ylabel('Total gastado promedio')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6,4))

evasion_tiempo_contrato.plot(kind='bar', color='b')

plt.title('Tiempo de contrato promedio según evasión')
plt.xlabel('Churn (0 = Permanece, 1 = Se dio de baja)')
plt.ylabel('Tenure promedio (meses)')
plt.xticks(rotation=0)
plt.show()
```

```python


plt.figure(figsize=(6,4))

plt.hist(datos[datos['Churn']==0]['Charges.Total'],
         bins=30,
         alpha=0.5,
         label='Permanece')

plt.hist(datos[datos['Churn']==1]['Charges.Total'],
         bins=30,
         alpha=0.5,
         label='Se dio de baja')

plt.title('Distribución del gasto total según evasión')
plt.xlabel('Total gastado')
plt.ylabel('Cantidad de clientes')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.hist(datos[datos['Churn']==0]['tenure'],
         bins=20,
         alpha=0.5,
         label='Permanece')

plt.hist(datos[datos['Churn']==1]['tenure'],
         bins=20,
         alpha=0.5,
         label='Se dio de baja')

plt.title('Distribución del tiempo de contrato según evasión')
plt.xlabel('Tenure (meses)')
plt.ylabel('Cantidad de clientes')
plt.legend()
plt.show()
```

<h2 style="color:#1C4587; font-family:Verdana;">📌 Introducción</h2>

<p style="font-family:Georgia; font-size:16px; line-height:1.6;">
Este informe presenta el análisis de la base de datos de clientes de la empresa de telecomunicaciones
<b>TelecomX LATAM</b>. El objetivo principal es comprender el fenómeno de
<b>Churn</b> (cancelación o evasión de clientes).
</p>

<p style="font-family:Georgia; font-size:16px;">
En el sector de telecomunicaciones, <b>retener clientes existentes es considerablemente más rentable
que adquirir nuevos usuarios</b>. Por esta razón, identificar los factores asociados al churn permite
desarrollar estrategias de retención más efectivas.
</p>

---

<h2 style="color:#134F5C; font-family:Verdana;">🧹 Limpieza y Tratamiento de Datos</h2>

<p style="font-family:Georgia; font-size:16px;">
Para garantizar la calidad del análisis, los datos en formato JSON fueron sometidos a un proceso
de <b>Extracción, Transformación y Carga (ETL)</b>.
</p>

<h3 style="color:#3D85C6;">Normalización de Datos</h3>

<ul style="font-family:Georgia; font-size:15px;">
<li>Expansión de las columnas <b>customer</b>, <b>phone</b>, <b>internet</b> y <b>account</b>.</li>
<li>Transformación de diccionarios en variables independientes.</li>
</ul>

<h3 style="color:#3D85C6;">Corrección de Datos</h3>

<ul style="font-family:Georgia; font-size:15px;">
<li>Detección de valores vacíos en <b>Charges.Total</b>.</li>
<li>Reemplazo de valores en blanco por <b>0</b>.</li>
<li>Conversión de la variable a formato numérico (<b>float</b>).</li>
</ul>

<h3 style="color:#3D85C6;">Estandarización</h3>

<ul style="font-family:Georgia; font-size:15px;">
<li>Variables categóricas (Sí/No) transformadas en valores <b>1 y 0</b>.</li>
<li>Facilita el análisis estadístico y el uso de modelos predictivos.</li>
</ul>

<h3 style="color:#3D85C6;">Ingeniería de Variables</h3>

<ul style="font-family:Georgia; font-size:15px;">
<li><b>Cuentas_Diarias</b> = Cargo mensual / 30</li>
<li><b>Servicios</b> = suma de servicios contratados por cliente</li>
</ul>

---

<h2 style="color:#6A329F; font-family:Verdana;">📊 Análisis Exploratorio de Datos</h2>

<p style="font-family:Georgia; font-size:16px;">
A través de diferentes visualizaciones se analizaron patrones asociados con la cancelación del servicio.
</p>

<h3 style="color:#674EA7;">Distribución de Churn</h3>

<p style="font-family:Georgia;">
Un gráfico circular permitió observar la proporción entre clientes que permanecen en la empresa
y aquellos que cancelan el servicio.
</p>

<h3 style="color:#674EA7;">Factores Demográficos</h3>

<ul style="font-family:Georgia;">
<li>El género no presenta diferencias significativas.</li>
<li>Los <b>adultos mayores</b> muestran mayor probabilidad de churn.</li>
<li>Clientes sin pareja ni dependientes presentan mayor cancelación.</li>
</ul>

<h3 style="color:#674EA7;">Tipo de Contrato</h3>

<p style="font-family:Georgia;">
Los contratos <b>Month-to-Month</b> presentan la mayor tasa de cancelación,
mientras que los contratos anuales muestran mayor estabilidad.
</p>

<h3 style="color:#674EA7;">Antigüedad del Cliente</h3>

<p style="font-family:Georgia;">
La mayoría de las cancelaciones ocurre durante los primeros meses de relación con la empresa.
</p>

<h3 style="color:#674EA7;">Número de Servicios</h3>

<p style="font-family:Georgia;">
Los clientes con <b>2 a 4 servicios</b> presentan los niveles más altos de evasión,
mientras que aquellos con múltiples servicios muestran mayor fidelidad.
</p>

---

<h2 style="color:#B45F06; font-family:Verdana;">📈 Conclusiones</h2>

<ul style="font-family:Georgia; font-size:16px;">
<li>El mayor riesgo de cancelación ocurre en los primeros meses del cliente.</li>
<li>Los contratos mensuales generan menor compromiso.</li>
<li>Los clientes con más servicios contratados presentan mayor fidelización.</li>
<li>El método de pago <b>Electronic Check</b> está asociado a mayores cancelaciones.</li>
</ul>

---

<h2 style="color:#38761D; font-family:Verdana;">💡 Recomendaciones</h2>

<ul style="font-family:Georgia; font-size:16px;">
<li>Promover contratos de largo plazo mediante incentivos.</li>
<li>Implementar programas de fidelización en los primeros meses.</li>
<li>Ofrecer promociones para aumentar el número de servicios contratados.</li>
<li>Fomentar métodos de pago automáticos.</li>
</ul>
